```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
Dataset = [['Milk', 'Onion', 'Bread', 'Butter', 'Eggs', 'Yogurt'],
           ['Cheese', 'Onion', 'Bread', 'Butter', 'Eggs', 'Yogurt'],
           ['Milk', 'Apple', 'Butter', 'Eggs'],
           ['Milk', 'Orange', 'Corn', 'Butter', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Butter', 'Ice cream', 'Eggs']]
```

```python
te=TransactionEncoder()
```

```python
data= te.fit(Dataset).transform(Dataset)
```

```python
import pandas as pd
```

```python
Productdata=pd.DataFrame(data,columns=te.columns_)
```

```python
Productdata
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

#Determining the frequent factors
frequent_factors = apriori(Productdata,use_colnames=True,min_support=0.1,max_len=4)

#Filtering the factors on the basis of conditions
rules = association_rules(frequent_factors,metric ='lift',min_threshold=1)

#Displaying the first 10 records
print(rules.head(10).sort_values(by='confidence',ascending =False))
```

```python
#List the products which were sold min 50% times
apriori(Productdata,use_colnames=True,min_support=0.5)
```

```python
#the data of the market basket analysis should be in the list format only
frequent_itemsets = apriori(Productdata,min_support=0.6,use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x:len(x))
```

```python
frequent_itemsets[(frequent_itemsets['length']>=2) & (frequent_itemsets['support']>=0.7)]
```

```python
frequent_itemsets[(frequent_itemsets['itemsets'] == {'Onion','Butter'})]
```

```python
target = '{\'Milk\'}'

#Searching required strings in consequents and sorting values on confidence
results_Milk = rules[rules['antecedents'].astype(str).str.contains(target,na =False)].sort_values(by='confidence',ascending =False)

results_Milk
```
